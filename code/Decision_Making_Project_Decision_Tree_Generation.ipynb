```python
import pandas as pd
import numpy as np
import tensorflow_decision_forests as tfdf
import keras
# Load TF-DF
import tensorflow_decision_forests as tfdf
import pandas as pd
```

# Loading the Generated Datasets

```python
Hellmuth_vs_Dwan_post_flop_dataset = pd.read_csv('Hellmuth_vs_Dwan_post_flop_dataset.csv')
Hellmuth_vs_Dwan_post_flop_dataset = Hellmuth_vs_Dwan_post_flop_dataset.iloc[: , 1:]
Hellmuth_vs_Dwan_pre_flop_dataset = pd.read_csv('Hellmuth_vs_Dwan_pre_flop_dataset.csv')
Hellmuth_vs_Dwan_pre_flop_dataset = Hellmuth_vs_Dwan_pre_flop_dataset.iloc[:,1:]
Hellmuth_vs_Negreanu_post_flop_dataset = pd.read_csv('Hellmuth_vs_Negreanu_post_flop_dataset.csv')
Hellmuth_vs_Negreanu_post_flop_dataset = Hellmuth_vs_Negreanu_post_flop_dataset.iloc[:,1:]
Hellmuth_vs_Negreanu_pre_flop_dataset = pd.read_csv('Hellmuth_vs_Negreanu_pre_flop_dataset.csv')
Hellmuth_vs_Negreanu_pre_flop_dataset = Hellmuth_vs_Negreanu_pre_flop_dataset.iloc[:,1:]
```

```python
Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset = pd.read_csv('Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset.csv')
Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset = Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset.iloc[:, 1:]
Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset = pd.read_csv('Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset.csv')
Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset = Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset.iloc[:, 1:]
```

# Convert the pandas dataframes to tensorflow datasets

```python
Hellmuth_vs_Dwan_post_flop_dataset_as_tensor = tfdf.keras.pd_dataframe_to_tf_dataset(Hellmuth_vs_Dwan_post_flop_dataset, label='In_or_Out_Decision')
Hellmuth_vs_Dwan_pre_flop_dataset_as_tensor = tfdf.keras.pd_dataframe_to_tf_dataset(Hellmuth_vs_Dwan_pre_flop_dataset, label="In_or_Out_Decision")
Hellmuth_vs_Negreanu_post_flop_dataset_as_tensor = tfdf.keras.pd_dataframe_to_tf_dataset(Hellmuth_vs_Negreanu_post_flop_dataset, label="In_or_Out_Decision")
Hellmuth_vs_Negreanu_pre_flop_dataset_as_tensor = tfdf.keras.pd_dataframe_to_tf_dataset(Hellmuth_vs_Negreanu_pre_flop_dataset, label="In_or_Out_Decision")
```

```python
Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset_as_tensor = tfdf.keras.pd_dataframe_to_tf_dataset(Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset, label='Check_Or_Raise_Decision')
Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset_as_tensor = tfdf.keras.pd_dataframe_to_tf_dataset(Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset, label='Check_Or_Raise_Decision')
```

# Creating the Trees: Stay in vs Get Out

## Hellmuth vs Negreanu, pre flop

```python
Hellmuth_vs_Negreanu_pre_flop_model = tfdf.keras.RandomForestModel()
Hellmuth_vs_Negreanu_pre_flop_model.fit(Hellmuth_vs_Negreanu_pre_flop_dataset_as_tensor)
tfdf.model_plotter.plot_model_in_colab(Hellmuth_vs_Negreanu_pre_flop_model)
```

## Hellmuth vs Dwan, pre flop

```python
Hellmuth_vs_Dwan_pre_flop_model = tfdf.keras.RandomForestModel()
Hellmuth_vs_Dwan_pre_flop_model.fit(Hellmuth_vs_Dwan_pre_flop_dataset_as_tensor)
tfdf.model_plotter.plot_model_in_colab(Hellmuth_vs_Dwan_pre_flop_model)
```

## Hellmuth vs Negreanu, post flop

```python
Hellmuth_vs_Negreanu_post_flop_model = tfdf.keras.RandomForestModel()
Hellmuth_vs_Negreanu_post_flop_model.fit(Hellmuth_vs_Negreanu_post_flop_dataset_as_tensor)
tfdf.model_plotter.plot_model_in_colab(Hellmuth_vs_Negreanu_post_flop_model)
```

## Hellmuth vs Dwan, post flop

```python
Hellmuth_vs_Dwan_post_flop_model = tfdf.keras.RandomForestModel()
Hellmuth_vs_Dwan_post_flop_model.fit(Hellmuth_vs_Dwan_post_flop_dataset_as_tensor)
tfdf.model_plotter.plot_model_in_colab(Hellmuth_vs_Dwan_post_flop_model)
```

# Creating the Trees: Check or Raise

```python
Hellmuth_vs_Negreanu_check_or_raise_model = tfdf.keras.RandomForestModel()
Hellmuth_vs_Negreanu_check_or_raise_model.fit(Hellmuth_vs_Negreanu_post_flops_impossible_to_fold_dataset_as_tensor)
tfdf.model_plotter.plot_model_in_colab(Hellmuth_vs_Negreanu_check_or_raise_model)
```

```python
Hellmuth_vs_Dwan_check_or_raise_model = tfdf.keras.RandomForestModel()
Hellmuth_vs_Dwan_check_or_raise_model.fit(Hellmuth_vs_Dwan_post_flops_impossible_to_fold_dataset_as_tensor)
tfdf.model_plotter.plot_model_in_colab(Hellmuth_vs_Dwan_check_or_raise_model)
```
